#Gradient Boosting and Churn analysis

```python
# needed packages
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
from copy import deepcopy
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.metrics import classification_report
from sklearn.learning_curve import validation_curve
```

```python
churn_df = pd.read_csv('data/churn.csv')
col_names = churn_df.columns.tolist()

for name in col_names:
    print(name)
```

```python
churn_df.head()
```

```python
churn_df['Churn?'] = churn_df['Churn?'] == "True."
labels = churn_df["Churn?"]
features = churn_df.drop(['State', 'Area Code', 'Phone', 'Churn?'], axis=1)
```

```python
features.head()
```

We need to change the 'yes' 'no' string values to boolean ones 

```python
yes_no_cols = ["Int'l Plan", "VMail Plan"]
features[yes_no_cols] = (features[yes_no_cols] == 'yes')
```

```python
features_names = features.columns
```

```python
print("Customers that churned : %d" %np.sum(labels == True))
print("Customers that did not churn : %d" %np.sum(labels == False))
```

```python
dataframe = deepcopy(features)
dataframe['churn'] = labels
dataframe.corr()["churn"]
```

```python
# Parameters :
depth = 6
alpha = 0.1
n_iter = 100
```

```python

gb = GradientBoostingClassifier(max_depth=depth,
                                learning_rate=alpha,
                                n_estimators=n_iter,
                                max_features='auto',
                                verbose=1)
```

### Meta parameters : 
**Max_depth**

The max_depth parameter controls the depth of the tree :
 
 - Deeper trees explained more variance at the expense of biais of course. 
 - The depth is directly correlated to the interaction between variables. A depth of 2 (called decision stump) deals only with main effect. 
 - Decision stumps maybe useful but it is pretty rare that high order interactions bring any values. 

Bottom line max_depth values between 4 and 8 give the best results in most cases. 

#### Regularization parameters: 
**n_estimators**
The n_estimators parameter control the number of boosting iterations. Each iterations reduce the training error leading to possible overfitting. This problem can be adress by monitoring the validation error

**learning_rate**
The learning rate $\alpha$ is another parameters that controls the scale at wich each tree contribute. The smaller the learning rate the higher the training risk for the same number of iteration. 

The regularization parameter interact with each other, smaller values of $\alpha$ leads to higher number of iterations. It appears that the best strategy is to choose a small value of $\alpha$ ($< 0.1$) and then choose the number of iterations accordingly. 

*ref : T. Hastie, R. Tibshirani and J. Friedman. Elements of Statistical Learning Ed. 2, Springer, 2009.*


## Methodology

We will follow a quite stringent methodology in order to choose the best model for our problem.

 - First we will split our data set into a training and a test set. We will keep the test set for the final evaluation of accuracy of our model. The test set should never be a mean to select the proper model but only to evaluate final accuracy.
 - we'll try to figure out which predictor are important 
 - We will then illustrate the overfitting problem mentionned above
 - then we will do a grid search on the n_estimators parameters to select the best model
 - finally we will assess the performace of our best estimator using the precision, recall, F1-score and support. 


```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, train_size=2400)
```

```python
gb.fit(X_train, y_train)
```

```python
# give explanations 
FIG_SIZE = (16,10)
importance_s = pd.Series(gb.feature_importances_, index=list(features_names))
importance_s /= importance_s.max()
importance_s.sort()
importance_s.plot(kind='barh', color='darkblue', figsize=FIG_SIZE)
```

Validation curve. A tool to monitor overfitting

```python
param_range = np.arange(20, 150, 10)
metric = 'accuracy'
training_scores, validation_scores = validation_curve(GradientBoostingClassifier(max_depth=6), 
                                                      X_train, y_train, 
                                                      "n_estimators", param_range,
                                                      n_jobs=-1,
                                                      cv=4, scoring=metric)
```

```python
training_scores_mean = np.mean(training_scores, axis=1)
training_scores_std = np.std(training_scores, axis=1)
validation_scores_mean = np.mean(validation_scores, axis=1)
validation_scores_std = np.std(validation_scores, axis=1)
plt.figure(num=None, figsize=FIG_SIZE, dpi=600, facecolor='w', edgecolor='k')
plt.title("Validation Curve with Gradient Boosting")
plt.xlabel("$Number of iterations$")
plt.ylabel("Accuracy")
plt.xlim(np.min(param_range), np.max(param_range))
plt.ylim(0.93, 1)
plt.semilogx(param_range, training_scores_mean, label="Training score", color="mediumblue")
plt.fill_between(param_range, training_scores_mean - training_scores_std,
                 training_scores_mean + training_scores_std, alpha=0.2, color="lightskyblue")
plt.semilogx(param_range, validation_scores_mean, label="validation score",
             color="coral")
plt.fill_between(param_range, validation_scores_mean - validation_scores_std,
                 validation_scores_mean + validation_scores_std, alpha=0.2, color="lightcoral")
plt.legend(loc="best")
```

```python
# grid search
param_grid = {'n_estimators': np.arange(70, 110, 5)}
grid_search = GridSearchCV(gb, param_grid=param_grid, cv=5, verbose=0, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
print("the winner is :", grid_search.best_params_)
print("with a score of : %f" % grid_search.best_score_)
plt.figure(figsize=FIG_SIZE, dpi=1200)
plt.plot([c.mean_validation_score for c in grid_search.grid_scores_], label="validation accuracy")
plt.xticks(np.arange(7), param_grid['n_estimators']); plt.xlabel("depth"); plt.ylabel("Accuracy");plt.legend(loc='best');
```

Here we select the model, now we will compute its accuracy on the test set that we kept for the end :

```python
gb_best = GradientBoostingClassifier(n_estimators=90, 
                                     max_depth=6, 
                                     max_features='auto', 
                                     learning_rate=0.1)
gb_best.fit(X_train, y_train)
y_pred = gb_best.predict(X_test)
proba_pred = gb_best.predict_proba(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
